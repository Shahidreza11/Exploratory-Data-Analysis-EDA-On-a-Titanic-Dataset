## Creating Family Type:

**Family type is defined as:**

- 1: Alone
- 2 to 4: Small family
- 3: Large family


```python
def familysize(num):
    if num==1:
        return 'alone'
    elif num>1 and num<5:
        return 'small'
    else:
        return 'large'
concated['family_type']=concated['Family_size'].apply(familysize)
```

```python
concated
```

```python
pd.crosstab(concated['Survived'],concated['family_type'])
```

```python
pd.crosstab(concated['Survived'],concated['family_type'],normalize='columns')*100
```

**Conclusion**
- Traveling alone on the Titanic results in a 69% chance of dying.
- Traveling in a large family increases the chance of dying to 83%.
- Traveling in a small family on the Titanic offers a higher chance (57%) of survival.
